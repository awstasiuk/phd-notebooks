```python
import nmresearch as nmr
import numpy as np
import matplotlib.pyplot as plt
```

```python
zz_echo=nmr.Experiment(221,folder='expt9').tpc(cycle=120,use_real=False)
cycle=120
t_list=np.array([k*cycle for k in range(len(zz_echo))])
```

```python
x = t_list
y = -np.log(zz_echo)

pz = np.polyfit(x,y, 1)
plt.plot(x, x*pz[0] + pz[1],label="fit")
plt.scatter(x,y,label="data")

plt.xlabel("time (us)")
plt.ylabel("minus log of signal")
plt.title("decay of z quadrature Loscmidt Echo")
plt.show()

print(1/pz[0])
```

```python
zz_echo = nmr.Experiment(232,folder='expt9').tpc(cycle=96,use_real=False,title="DQ Echo, c1=0.05")
cycle=24*4
t_list=np.array([k*cycle for k in range(len(zz_echo))])
```
