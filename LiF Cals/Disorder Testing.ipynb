```python
import nmresearch as nmr
import numpy as np
import matplotlib.pyplot as plt
```

```python
cycle60 = nmr.Experiment(101,folder='expt9').tpc(cycle=60)
t_list = np.array([i*60 for i in range(len(cycle60))])
```

```python
start = 8
end = -25

x = t_list[start:end]
y = -np.log(cycle60[start:end])

p = np.polyfit(x,y, 1)
plt.plot(x, x*p[0] + p[1],label="fit")
plt.scatter(x,y,label="data")

plt.xlabel("time (us)")
plt.ylabel("minus log of signal")

plt.show()

print(1/p[0])
```

```python
plt.scatter(t_list,cycle60,label="data")
plt.scatter(t_list,np.exp(-t_list*p[0]),label="exponential fit, late time")
plt.scatter(t_list, cycle60/np.exp(-t_list*p[0]),label="renormalized data")
plt.ylim([-.1,1.1])
plt.legend()
plt.show()
```
