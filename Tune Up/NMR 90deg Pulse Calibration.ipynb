```python
file = path+"\\121"
print(file)
```

```python
dic, arr = ng.fileio.bruker.read(file)
```

```python
vals = arr[:,:,0]
```

```python
plt.plot(np.imag(vals[0]), label="90 deg")
plt.plot(np.imag(vals[1]), label="90+360 deg")
plt.plot(np.imag(vals[2]), label="90+720 deg")
plt.title("Bloch Sphere Wrapping Signal")
plt.legend()
plt.show()

high_contrast = np.imag(vals[0])*np.imag(vals[1])*np.imag(vals[2])
plt.plot(high_contrast)
plt.title("High contrast signal plot")
plt.show()

print("array index of max signal is " + str(np.argmax(high_contrast)))
```
