```python
"""
Calibrate the T1 using a population inversion & recovery experiment.

This is implemented experimentally using t1_calibration
"""

sig_raw = nmr.Experiment(96,'expt9').load_tpc(normalize=False)
sig = sig_raw/abs(sig_raw[0])
tau = np.array([0.000005, 0.000050, 1.00000, 5.00000, 10.00000, 15.0000, 20.00000, 25.00000, 30.00000, 60])

p = np.polyfit(tau,-np.log((1-sig)/2), 1)
plt.plot(tau, tau*p[0] + p[1])
plt.scatter(tau,-np.log((1-sig)/2))
plt.show()

plt.scatter(tau,sig)
plt.plot(tau, 1-2*np.exp(-p[1])*np.exp(-tau*p[0]), label=1/p[0])
plt.legend()
plt.xlabel("Inversion recovery time (s)")
plt.ylabel("Normalized Signal")
plt.show()

print(5/p[0])
```

```python
"""
; For DQ Loschmidt echo expts

define delay tau1
"tau1=d3-p1/2"
define delay tau2
"tau2=d3*(1-cnst1)-p1"
define delay tau3
"tau3=2*d3*(1+cnst1)-p1"

define delay tau1r
"tau1r=d3-p1/2"
define delay tau2r
"tau2r=d3*(1+cnst1)-p1"
define delay tau3r
"tau3r=2*d3*(1-cnst1)-p1"
"""
def to_str(lst):
    return ["{:10.8f}".format(val) for val in lst]

cnst1=[0.0, .02, .04, .06, .08, .10]

us = 10**-6
d3=5*us
p1=1.02*us
tau1 = [d3-p1/2 for c1 in cnst1]
tau2 = [d3*(1-c1)-p1 for c1 in cnst1]
tau3 = [2*d3*(1+c1)-p1 for c1 in cnst1]
tau1r = [d3-p1/2 for c1 in cnst1]
tau2r = [d3*(1+c1)-p1 for c1 in cnst1]
tau3r = [2*d3*(1-c1)-p1 for c1 in cnst1]

str_list = []

str_list.append("define list<delay> tau1 = {" + " ".join(to_str(tau1)) + "}")
str_list.append("define list<delay> tau2 = {" + " ".join(to_str(tau2)) + "}")
str_list.append("define list<delay> tau3 = {" + " ".join(to_str(tau3)) + "}")

str_list.append("define list<delay> tau1r = {" + " ".join(to_str(tau1r)) + "}")
str_list.append("define list<delay> tau2r = {" + " ".join(to_str(tau2r)) + "}")
str_list.append("define list<delay> tau3r = {" + " ".join(to_str(tau3r)) + "}")

print(f"\n".join(str_list))
```
