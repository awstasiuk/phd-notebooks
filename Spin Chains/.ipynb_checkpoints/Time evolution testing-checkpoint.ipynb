```python
import numpy as np
import fermion as f
import matplotlib.pyplot as plt
from pandas import DataFrame
from fermion import Math as fm
```

```python
n=30
```

```python
dq = f.Operator.double_quantum(n)
Z = f.Operator.global_Z(n)


H = dq
```

```python
time_evo = f.Unitary(H,.05,2.05)
```

```python
zz = time_evo.global_zz()
```

```python
plt.plot(time_evo.t, zz)
```

```python
plt.plot(time_evo.t, time_evo.local_zz(0,1))
```

```python
Zt = [time_evo.evolve_op(Z,t) for t in time_evo.t]
```

```python
comm = [Zt[i].commutator(Z).normal_order().compactify() ** 2 for i in range(len(Zt))]
```

```python
signal = [-comm[i].trace()/8 for i in range(len(Zt))]
```

```python
plt.plot(time_evo.t, signal)
```
