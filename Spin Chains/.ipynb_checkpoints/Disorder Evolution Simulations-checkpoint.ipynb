```python
import numpy as np
import nmresearch as nmr
from nmresearch import fermion as f
import matplotlib.pyplot as plt
from nmresearch.fermion import Math as fm

from timeit import default_timer as timer
```

```python
n=50
```

```python
disorder = [np.random.normal(0,.1) for _ in range(n)]
rZ = f.Operator.disorder_Z(n, disorder)
dq = nmr.Operator.double_quantum(n,periodic=False)
Z = nmr.Operator.global_Z(n)
H= dq

time_evo = f.Unitary(H,.05,4.05)
```

```python
start=timer()
tpc = [time_evo.local_zz(n//2, r) for r in range(n)]
end=timer()
print(end-start)
```

```python
test = np.array(tpc)
```

```python
test.shape
```

```python
fig, ax = plt.subplots()

c = ax.pcolormesh(time_evo.t, list(range(-25,25,1)), tpc)
ax.set_title(r'$<Z_0 Z_r(t)>$, $h/J=0.0$')
# set the limits of the plot to the limits of the data
ax.axis([0,4, -25, 24])
ax.set_ylabel('r')
ax.set_xlabel('J t')
fig.colorbar(c, ax=ax)
plt.savefig("dq_tpc_h0.png",format='png',dpi=900,bbox_inches='tight')
plt.show()
```

```python
zz_tpc_avg = np.zeros((n,len(time_evo.t)))
```

```python
zz_tpc_avg.shape
```

```python
start = timer()

trials = 10000
zz_tpc_avg = np.zeros((n,len(time_evo.t)))
for _ in range(trials):
    disorder = [np.random.normal(0,.1) for _ in range(n)]
    rZ = f.Operator.disorder_Z(n, disorder)

    H=rZ+dq

    time_evo = f.Unitary(H,.05,4.05)
    
    zz_tpc_avg = zz_tpc_avg + np.array([time_evo.local_zz(n//2, r) for r in range(n)])/(trials)
    
end = timer()
print(end-start)

fig, ax = plt.subplots()

c = ax.pcolormesh(time_evo.t, list(range(-25,25,1)), zz_tpc_avg)
ax.set_title(r'$<Z_0 Z_r(t)>$, $h/J=0.1$')
# set the limits of the plot to the limits of the data
ax.axis([0,4, -25, 24])
ax.set_ylabel('r')
ax.set_xlabel('J t')
fig.colorbar(c, ax=ax)
plt.savefig("dq_tpc_h1.png",format='png',dpi=900,bbox_inches='tight')
plt.show()
```

```python
start = timer()

trials = 10000
zz_tpc_avg = np.zeros((n,len(time_evo.t)))
for _ in range(trials):
    disorder = [np.random.normal(0,.5) for _ in range(n)]
    rZ = f.Operator.disorder_Z(n, disorder)

    H=rZ+dq

    time_evo = f.Unitary(H,.05,4.05)
    
    zz_tpc_avg = zz_tpc_avg + np.array([time_evo.local_zz(n//2, r) for r in range(n)])/(trials)
    
end = timer()
print(end-start)
fig, ax = plt.subplots()

c = ax.pcolormesh(time_evo.t, list(range(-25,25,1)), zz_tpc_avg)
ax.set_title(r'$<Z_0 Z_r(t)>$, $h/J=0.5$')
# set the limits of the plot to the limits of the data
ax.axis([0,4, -25, 24])
ax.set_ylabel('r')
ax.set_xlabel('J t')
fig.colorbar(c, ax=ax)
plt.savefig("dq_tpc_h5.png",format='png',dpi=900,bbox_inches='tight')
plt.show()
```
