```python
import numpy as np
import nmresearch as nmr
import matplotlib.pyplot as plt
from pfapack import pfaffian as pfa
import scipy as sp

from timeit import default_timer as timer
```

```python
n=30
```

```python
dq = nmr.Operator.double_quantum(n)
Z = nmr.Operator.global_Z(n)
time_evo = nmr.Unitary(dq,.05, 2.05)
Zt = [time_evo.evolve_op(Z,t) for t in time_evo.t]
```

```python
xx_corr = np.zeros(len(time_evo.t))
center=int(n/2)
for a in range(n):
    p = nmr.PauliString(["X","X"],[center,a],[True, False])
    xx_corr = [prev+cur for prev,cur in zip(xx_corr, time_evo.pauli_string_expectation(p))]
```

```python
plt.plot(time_evo.t, xx_corr)
```

```python
plt.plot(time_evo.t, [np.sqrt(-np.log(val)) for val in xx_corr])
plt.plot(time_evo.t, [2*t for t in time_evo.t])
plt.show()
```
