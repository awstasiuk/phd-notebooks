```python
import nmresearch as nmr
import matplotlib.pyplot as plt
```

```python
n=30
```

```python
dq = nmr.Operator.double_quantum(n)
```

```python
time_evo = nmr.Unitary(dq,.05, 2.05)
Z = nmr.Operator.global_Z(n)
Zt = [time_evo.evolve_op(Z,t) for t in time_evo.t]
```

```python
from timeit import default_timer as timer
```

```python
start = timer()
xx_otoc = time_evo.global_OTOC("X","X")
end = timer()
print(end-start)
```

```python
plt.plot(time_evo.t,xx_otoc)
plt.xlabel("J t")
plt.ylabel("Normalized Signal")
plt.title("<|[X(t),X]|^2>, N=30, H=DQ")
plt.show()
```

```python
start = timer()
xz_otoc = time_evo.global_OTOC("X","Z")
end = timer()
print(end-start)
```

```python
plt.plot(time_evo.t,xz_otoc)
plt.xlabel("J t")
plt.ylabel("Normalized Signal")
plt.title("<|[X(t),Z]|^2>, N=30, H=DQ")
plt.show()
```

```python
start = timer()
zx_otoc = time_evo.global_OTOC("Z","X")
end = timer()
print(end-start)
```

```python
plt.plot(time_evo.t,zx_otoc)
plt.xlabel("J t")
plt.ylabel("Normalized Signal")
plt.title("<|[Z(t),X]|^2>, N=30, H=DQ")
plt.show()
```
