```python
import nmresearch as nmr
from timeit import default_timer as timer
import matplotlib.pyplot as plt
```

```python
n=6
```

```python
dq = nmr.Operator.double_quantum(n)
Z = nmr.Operator.global_Z(n)
time_evo = nmr.Unitary(dq,.05, 2.05)
Zt = [time_evo.evolve_op(Z,t) for t in time_evo.t]
```

```python
start = timer()
zz_otoc_sym = time_evo.global_OTOC("Z","Z")
zz_otoc_full = time_evo.global_OTOC("Z","Z", use_true=True)
end = timer()
print(end-start)
```

```python
plt.plot(time_evo.t, zz_otoc_sym,'r')
plt.plot(time_evo.t, zz_otoc_full,'g')
plt.show()
```

```python
start = timer()
xx_otoc_sym = time_evo.global_OTOC("X","X")
xx_otoc_full = time_evo.global_OTOC("X","X", use_true=True)
end = timer()
print(end-start)
```

```python
plt.plot(time_evo.t, xx_otoc_sym, time_evo.t, xx_otoc_full)
```
