```python
import nmresearch as nmr
import numpy as np
import matplotlib.pyplot as plt
```

```python
#checking signal magnitude at T=5 Cycles for variable c2
expt = nmr.Experiment(14, folder='expt10')
data = expt.load_tpc(normalize=False,use_real=True)
c2_lst = np.arange(-.05,.05,.01)
plt.scatter(c2_lst, data, label='c2 values')
plt.show()
```

```python
#time domain fid with ken, tau=4us
# c2=.05
expt = nmr.Experiment(15, folder='expt10')
data = expt.load_tpc(normalize=False,use_real=True)
T = 4*24
times = np.arange(0, 25*T, T)
plt.scatter(times, data, label='tau=4us, c2=0.05')

# time domain fid with ken, tau=5us
# c2=.05
expt = nmr.Experiment(16, folder='expt10')
data = expt.load_tpc(normalize=False,use_real=True)
T = 5*24
times = np.arange(0, 25*T, T)
plt.scatter(times, data, label='tau=5us, c2=0.05')


plt.legend()
plt.xlabel('Clock Time (us)')
plt.ylabel('Signal')
plt.show()
```
