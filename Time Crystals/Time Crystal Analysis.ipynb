```python
import nmresearch as nmr
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
vals = nmr.Experiment(140).load_tpc3d()
```

```python
kicks = list(range(50))
max(vals[0])
```

```python
plt.scatter(kicks,vals[0])
plt.xlabel("Number of Cycles")
plt.ylabel("(Absolute) Signal Intensity")
plt.title("Spin chain w/ disorder under periodic driving")
plt.show()


amp = sp.fft.fft(vals[0])
freq = [i*(1/(len(kicks))) for i in range(len(kicks))]
plt.axvline(x=.5, ymin=0, color='red', linestyle='dotted', linewidth=1)
plt.plot(freq, abs(amp))
plt.show()
```

```python
vals = nmr.Experiment(156).load_tpc3d()
kicks = list(range(vals.shape[1]))

avg = [np.mean(vals[:,i]) for i in range(vals.shape[1])]
ci_95 = [1.96*np.std(vals[:,i]) for i in range(vals.shape[1])]

plt.scatter(kicks,avg)
plt.errorbar(kicks, avg, yerr=ci_95, fmt="o")
plt.xlabel("Number of Cycles")
plt.ylabel("(Absolute) Signal Intensity")
plt.title("Spin chain w/ disorder under periodic driving")
plt.show()


amp = sp.fft.fft(vals[0])
freq = [i*(1/(len(kicks))) for i in range(len(kicks))]
plt.axvline(x=.5, ymin=0, color='red', linestyle='dotted', linewidth=1)
plt.plot(freq,abs(amp))
plt.show()
```
