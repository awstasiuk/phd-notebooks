```python
import nmresearch as nmr
import numpy as np
import matplotlib.pyplot as plt
```

```python
ham = nmr.Hamiltonian(12)
```

```python
ham.H_cs_scrambling()
```

```python
operator =  nmr.PauliMatrix(12)
sigz = (1/2**6)*nmr.lanczos.op_basis.vec(operator.sigmaZ(0))
```

```python
test = ham.to_super()
```

```python
test
```

```python
(1124057088 * 64 * 2 / 8)/1000000000
```

```python
solver = nmr.Lanczos(test,sigz)
```

```python
solver.compute_lanczos_fast()
```

```python
signal_fast = solver.auto_correlation(times=np.arange(0,5,.1))
```

```python
plt.plot(np.arange(0,5,.1),signal_fast)
plt.xlabel("time")
plt.ylabel("sigma-z auto-correlation")
plt.show()
```
