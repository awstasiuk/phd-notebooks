```python
import nmresearch as nmr
import numpy as np
import matplotlib.pyplot as plt
```

```python
ham = nmr.Hamiltonian(10)
```

```python
ham.H_cs_scrambling()
```

```python
operator =  nmr.PauliMatrix(10)
sigz = (1/2**5)*nmr.lanczos.op_basis.vec(operator.sigmaZ(0))
```

```python
solver = nmr.Lanczos(ham.to_super(),sigz)
```

```python
solver.compute_lanczos_FRO(200)
```

```python
signal_FRO= solver.auto_correlation(times=np.arange(0,5,.1))
```

```python
plt.plot(np.arange(0,5,.1),signal_FRO)
plt.hlines(0.25,0,5,'k','dashed', label="y=0.25 assymptote")
plt.xlabel("time")
plt.ylabel("sigma-z auto-correlation")
plt.show()
```

```python
plt.scatter(list(range(len(solver.lanczos_coef))),solver.lanczos_coef)
plt.show()
```
