```python
import nmresearch as nmr
```

```python
N = 10
ham = nmr.Hamiltonian(N)
ham.H_cs_scrambling()
operator =  nmr.PauliMatrix(N)
sigz = (1/2**(N/2))*nmr.lanczos.op_basis.vec(operator.sigmaZ(0)).tocsc()
```

```python
superop = ham.to_super()
```

```python
solver = nmr.Lanczos(superop,sigz)
```

```python
import cProfile
```

```python
with cProfile.Profile() as pr:
    solver.compute_lanczos_fast()
    pr.print_stats(sort=1)
```

```python
temp = sigz.todense()
sigz_col = sigz.tocsc()
```

```python
superop = superop
supercsc = superop.tocsc()
```

```python
def test_csr():
    superop @ temp

def test_csc():
    supercsc @ temp

def test_sprse():
    superop @ sigz

def test_sprse_csc():
    superop.dot(sigz_col)
```

```python
timeit.timeit('test_csr()',number=10,globals=globals())
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time_vec = np.arange(0,5,.05)
```

```python
alph=.05
for N, data in sigs1.items():
    plt.plot(time_vec, data[0],label="N = " + str(N) + ", 300 trials")
    plt.fill_between(time_vec, data[0]-data[1], data[0]+data[1],alpha=alph)
    alph += 0.05
plt.hlines(0.25,2,5,'k','dashed', label=r"$<\sigma_z(t)\sigma_z> = 0.25$")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel(r"Evolution Time, $J_{max} t$")
plt.ylabel("NV auto-correlation")
plt.savefig("central_spin_ac_sims.pdf", format='pdf', dpi=900,bbox_inches='tight')
plt.show()
```

```python
means = np.zeros(12-4+1)
errs = np.zeros(12-4+1)
idx=0
for N, data in sigs1.items():
    means[idx] = np.mean(data[0])
    errs[idx] = np.sqrt(np.mean(data[1]**2))
    idx += 1
```

```python
ns = np.arange(4,13,1)
```

```python
plt.scatter(ns,means,label="Fitted Asumptotes")
#plt.plot(ns, 1/ns, label="1/N",color="k",linestyle="--")
plt.plot(ns, 1/ns+.16, label="y=1/N+0.16",color="r",linestyle="--")
plt.hlines(0.25,4,12,'k','dashed', label="y=0.25")
plt.errorbar(ns, means,errs)
plt.xlabel("$N$-Spin Simulated Asymptote")
plt.ylabel("Late time mean auto-correlation")
plt.legend()
plt.show()
```
