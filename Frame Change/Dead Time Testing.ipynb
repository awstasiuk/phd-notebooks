```python
def t_list(DE,n_points=5,imag=True):
    return [DE+3.35*(2*i+imag) for i in range(n_points)]
```

```python
delay_list = [5,10,15,20,25]

dead_list = [5,7.5,10,12.5,15]

data5 = nmr.Experiment(171,folder='expt9').nmr_data
data7_5 = nmr.Experiment(172,folder='expt9').nmr_data
data10 = nmr.Experiment(173,folder='expt9').nmr_data
data12_5 = nmr.Experiment(174,folder='expt9').nmr_data
data15 = nmr.Experiment(175,folder='expt9').nmr_data

data = np.imag(np.array([data5, data7_5, data10, data12_5, data15]))
```

```python
# delay 5
n_points=7

fid5 = []
tau5 = []

for idx,DE in enumerate(dead_list):
    plt.scatter(t_list(DE,n_points),data[idx,0,0:n_points],label=str(DE)+'us')
    fid5.extend(data[idx,0,0:n_points])
    tau5.extend(t_list(DE,n_points))
    
plt.title("Solid Echo - 5us")
plt.xlabel("Time since last pulse")
plt.ylabel("Signal (a.u.)")
plt.legend()
plt.show()
```
